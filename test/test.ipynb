```python
import torch
import open3d as o3d
import numpy as np
import spseg
```

```python
mesh = o3d.io.read_triangle_mesh("./pointcloud.ply")
vertices = torch.from_numpy(np.array(mesh.vertices).astype(np.float32))
faces = torch.from_numpy(np.array(mesh.triangles).astype(np.int64))
```

scannet [segmentator](https://github.com/Karbo123/segmentator/tree/master)

```python
superpoint = spseg.scannet.segment_mesh(vertices, faces)
superpoint.shape
```
